### **2.Diabetes Dataset EDA:**

````python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
````

````python
df=pd.read_csv('C:/Users/HP/Downloads/diabetes_model.csv')
df
````

````python
#Analyze individual variables (e.g., histograms, box plots)
df['SkinThickness'].hist(bins=25,color='lightgreen',edgecolor='black')
plt.title('Hist of SkinThickness')
plt.xlabel('SkinThickness')
plt.ylabel('Thickness')
plt.show()
sns.boxplot(x=df['SkinThickness'],color='skyblue')
plt.title('boxplot of SkinThickness')
plt.show()
````

````python
#Explore relationships between pairs of variables (e.g., scatter plots, correlation matrices)
````

````python
x=df['Insulin']
y=df['Age']
plt.scatter(x,y,alpha=0.4,edgecolors='black')
plt.title('Scatter plot:Insulin vs  Age',fontsize=16)
plt.xlabel('Insulin')
plt.ylabel('Age')
plt.show()
correlation = df.corr()
sns.heatmap(correlation,annot=True,cmap='pink',fmt='.3f')
plt.title('correlation matrix')
plt.show()
````

````python
##Investigate interactions among multiple variables (e.g., pair plots, 3D plots).
sns.pairplot(df,hue='Outcome',palette='husl')
plt.show()
#3d plot
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
scatter=ax.scatter(df['Pregnancies'],df['BloodPressure'],df['Age'],cmap='viridis')
ax.set_xlabel('Pregnancies')
ax.set_ylabel('BloodPressure')
ax.set_zlabel('Age')
plt.show()
````
